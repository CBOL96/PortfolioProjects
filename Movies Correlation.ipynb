```python
# Import packages
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize']=(12,8)#Adjust the configuration of the plot we will create
```

```python
#Read the data
df=pd.read_csv(r'C:\Users\carli\Documents\Data Analysis\Projects\Correlation Project\movies.csv')
df.head()
```

```python
#Looking for missing data
missing_data = df.isnull()
print(missing_data.head())
for column in missing_data.columns.values.tolist():
    per_missing = np.mean(df[column].isnull())*100
    print (missing_data[column].value_counts())
    print ('Percentage: ',per_missing,'%')
    print("")
```

```python
#Drop Missing Data
df = df.dropna()
```

```python
#Check the datatypes
df.dtypes
```

```python
# Chage Data types of columns
df['budget']=df['budget'].astype('int64')
df['gross']=df['gross'].astype('int64')
```

```python
# Año Corregido
df['year_correct'] = df['released'].str.extract(r'(\d{4})').astype(int)
df
```

```python
# Ordering by Gross Revenue
df.sort_values(by=['gross'], inplace=False,ascending=False)
```

```python
#looking for correlation

#Budget vs gross revenue

plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Gross Earnings vs Budget')
plt.xlabel('Budget')
plt.ylabel('Gross earnings')
plt.show()
```

```python
#plot Gross vs Budget using seaborn
sns.regplot(x='budget',y='gross',data=df,scatter_kws={"color":"red"},line_kws={"color":"blue"})
plt.show()
```

```python
df_numeric = df.select_dtypes(include=['number'])
df_numeric= df_numeric.drop('year', axis=1)
correlation_matrix = df_numeric.corr()
print(correlation_matrix)
```

```python
sns.heatmap(correlation_matrix,annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
#Assign numeric values to object types
df_numerized = df
for column in df_numerized.columns:
    if (df_numerized[column].dtype == 'object'):
        df_numerized[column]=df_numerized[column].astype('category')
        df_numerized[column]=df_numerized[column].cat.codes
df_numerized
```

```python
correlation_matrix = df_numerized.corr()
correlation_matrix=correlation_matrix.drop(["year","released"],axis=1)
sns.heatmap(correlation_matrix,annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```
